```python
aves_neo = []
neo=neo.split('\n')
order = ''
for a in neo:
    if a[:5]=='Order':
        order=a.replace(':',' ').split(' ')[1]
        pass
    if a=='':
        order=None
        pass
    if 'Family' in a:
        fam=a.strip().replace(':',' ').split(' ')
        aves_neo.append({'Order':order,fam[0]:fam[1]})
        
print(aves_neo)
```

```python
aves_para = []
para=para.split('\n')
order = ''
for a in para:
    if a[:5]=='Order':
        order=a.replace(':',' ').split(' ')[1]
        pass
    if a=='':
        order=None
        pass
    if 'Family' in a:
        fam=a.strip().replace(':',' ').split(' ')
        aves_para.append({'Order':order,fam[0]:fam[1]})
        
print(aves_para)
```

```python
neo = pd.DataFrame(aves_neo)
neo['subclass'] = 'Neognathae'
neo
```

```python
pala = pd.DataFrame(aves_para)
pala['subclass'] = 'Palaeognathae'
pala
```

```python
aves = pd.concat([neo,pala],ignore_index=True)
aves['class'] = 'Aves'
aves = aves.rename(columns={'Order':'order','Family':'family'})
aves
```
